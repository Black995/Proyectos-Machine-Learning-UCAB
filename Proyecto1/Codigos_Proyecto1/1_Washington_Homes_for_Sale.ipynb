### Feature engineering, crear atributos combinados

```python
def feature_engineering(data):
    data["bedrooms_per_floors"]=data["bedrooms"]/data["floors"]
    data["bathrooms_per_floors"]=data["bathrooms"]/data["floors"]

    return data
```

```python
#homes_data_copy = homes_data.copy()
homes_feature_engineered = feature_engineering(homes_data_copy)
homes_feature_engineered
```

### Transformación personalizada

```python
from sklearn.base import BaseEstimator, TransformerMixin

col_names = "bedrooms", "bathrooms", "floors"
bedrooms_ix, bathrooms_ix, floors_ix = [homes_data_copy.columns.get_loc(c) for c in col_names]

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_floors=True):
        self.add_bedrooms_per_floors = add_bedrooms_per_floors
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        bathrooms_per_floors = X[:, bathrooms_ix] / X[:, floors_ix]
        if self.add_bedrooms_per_floors:
            bedrooms_per_floors = X[:, bedrooms_ix] / X[:, floors_ix]
            return np.c_[X, bathrooms_per_floors, bedrooms_per_floors]
        else:
            return np.c_[X, bathrooms_per_floors]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_floors=False)
homes_extra_attribs = attr_adder.transform(homes_data_copy.values)

homes_extra_attribs = pd.DataFrame(
    homes_extra_attribs,
    columns=list(homes_data_copy.columns)+["bathrooms_per_floors"],
    index=homes_data_copy.index)
homes_extra_attribs.head()
```

### Pipelines de transformación

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('attribs_adder', CombinedAttributesAdder()),
    ('std_scaler', StandardScaler()),
])

homes_num_tr = num_pipeline.fit_transform(homes_num)
homes_num_tr
```

```python
from sklearn.compose import ColumnTransformer

num_attribs = list(homes_num)

full_pipeline = ColumnTransformer([
    ("num", num_pipeline, num_attribs),
])

homes_prepared = full_pipeline.fit_transform(homes_data_copy)
homes_prepared
```

### Validación cruzada (Regresión Lineal)

```python
def display_scores(scores):
    print("Scores:", scores)
    print("Mean:", scores.mean())
    print("Standard deviation:", scores.std())
```

```python
from sklearn.model_selection import cross_val_score

lin_scores = cross_val_score(model_lin_reg, X_test, y_test, scoring="neg_mean_squared_error", cv=10)
lin_rmse_scores = np.sqrt(-lin_scores)

display_scores(lin_rmse_scores)
```
